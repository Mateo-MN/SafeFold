```python
import json
import pandas as pd

def load_json(path):
    rows = []

    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                obj = json.loads(line)
            except json.JSONDecodeError:
                continue

            row = {"uniprot_id": obj["uniprot_id"]}
            row.update(obj["GO_terms"].get("0", {}))
            rows.append(row)

    return pd.DataFrame(rows)

df_toxic = load_json("./go_terms_toxic.json")
df_non_toxic = load_json("./go_terms_not_toxic.json")
```
